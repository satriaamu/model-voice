```python
# batch processing
from pydub import AudioSegment
import os

# Replace with your input and output directories
input_directory = '/content/drive/MyDrive/dataset/input/ryan wav'
output_directory = '/content/drive/MyDrive/dataset/input/mono audio/ref ryan'

# List all files in the input directory
audio_files = os.listdir(input_directory)

# Iterate through each audio file
for file_name in audio_files:
    if file_name.endswith('.wav'):
        input_path = os.path.join(input_directory, file_name)
        output_path = os.path.join(output_directory, file_name.replace('.wav', '-mono.wav'))

        # Load the audio file
        sound = AudioSegment.from_wav(input_path)

        # Convert stereo to mono
        sound = sound.set_channels(1)

        # Export the mono audio to the output directory
        sound.export(output_path, format="wav")

        print(f"Converted {file_name} to mono and saved to {output_path}")
```
